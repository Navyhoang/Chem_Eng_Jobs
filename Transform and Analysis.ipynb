# Data Tranformation

```python
# Dependencies
import pandas as pd 
import os 
import numpy as np
```

```python
filepath = f"CA_chemengjobs.csv"
df = pd.read_csv(filepath, index_col=0)
df.head()
```

```python
# Remove the Link column
df = df[["Job Title Index", "Job ID", "Job Title", "Company Name", "Company Location"]]

df.head()
```

```python
# Check if there is any is null in the dataframe
df.isnull().sum()
```

```python
# Find unique job titles and count
df["Job Title"].value_counts()
```

```python
# Group by location and count
df_location = df.groupby(["Company Location"]).count()
df_location = df_location[["Job ID"]].sort_values(["Job ID"], ascending=False)
df_location.head(50)
```

```python
# Filter job titles that has string "Junior" 

df_junior = df[df['Job Title'].str.contains('Junior', na=False, regex=False)]
df_junior
```

```python
# Find unique job titles and count
df_junior["Job Title"].value_counts()
```

```python
# Group by location and count
df_junior_location = df_junior.groupby(["Company Location"]).count()
df_junior_location = df_junior_location[["Job ID"]].sort_values(["Job ID"], ascending=False)
df_junior_location.head(50)
```

```python
# Filter job titles that has string "Jr" 

df_jr = df[df['Job Title'].str.contains('Jr', na=False, regex=False)]
df_jr
```

```python
# Find unique job titles and count
df_jr["Job Title"].value_counts()
```

```python
# Group by location and count
df_jr_location = df_jr.groupby(["Company Location"]).count()
df_jr_location = df_jr_location[["Job ID"]].sort_values(["Job ID"], ascending=False)
df_jr_location.head(50)
```
